```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("Translated_London_Hotel_review_Dataset.csv")
df.head()
```

```python
def label_reviews(rating):

  '''Label the Review ratings in to positive, negative and neutral'''
  
  if rating <= 2:
    return "neg"
  elif rating == 3:
    return "neu"
  else:
    return "pos"
```

```python
df['label'] = df['Review Rating'].apply(label_reviews)
```

```python
sentiment_count = df["label"].value_counts()
plt.pie(sentiment_count,
         labels=sentiment_count.index,
         autopct='%1.1f%%',
         shadow=True,
         startangle=140)
plt.show()
```

```python
print(f'Positive reviews: {df[df["label"] == "pos"].count()[0]}')
print(f'Negative reviews: {df[df["label"] == "neg"].count()[0]}')
print(f'Negative reviews: {df[df["label"] == "neu"].count()[0]}')
```

```python
pos_reviews = df[df["label"] == "pos"]
txt = " ".join( review.lower() for review in pos_reviews['Translated Review'])
wordcloud = WordCloud().generate(txt)
plt.figure( figsize=(15,10) )
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title('Word Cloud for postive reviews')
plt.show()
```

```python
neg_reviews = df[df["label"] == "neg"]
txt = " ".join( review.lower() for review in neg_reviews['Translated Review'])
wordcloud = WordCloud().generate(txt)
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title('Word Cloud for negative reviews')
plt.show()
```

```python
neu_reviews = df[df["label"] == "neu"]
txt = " ".join( review.lower() for review in neu_reviews['Translated Review'])
wordcloud = WordCloud().generate(txt)
plt.figure( figsize=(15,10) )
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title('Word Cloud for neutral reviews')
plt.show()
```
